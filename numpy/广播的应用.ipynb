## 2.画一个二位函数

```python
x = np.linspace(0,5,500)
y = np.linspace(1,5,500)[:,np.newaxis]

z=np.sin(x)**10+np.cos(10+y*x)*np.cos(x)
```

```python
plt.imshow(z,origin='lower',extent=[0,5,0,5],
          cmap='viridis')
plt.colorbar();
```
